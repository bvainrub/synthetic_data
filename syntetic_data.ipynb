```python
# Loading CSV file

Ecommerce = pd.read_csv("/workspaces/synthetic_data/e-commerce.csv.csv")

Ecommerce.head()
```

```python
from faker import Faker
import random

def anonymous(df):
    local = {
        "United Kingdom": "en_GB",
        "France": "fr_FR",
        "Australia": "en_AU",
        "Netherlands": "nl_NL",
        "Germany": "de_DE",
        "Norway": "no_NO",
        "EIRE": "ga_IE",
    }

    # Rough bounding boxes (lat/lon) for realism
    country_coords = {
        "United Kingdom": (49.9, 58.7, -8.6, 1.8),
        "France": (41.3, 51.1, -5.1, 9.6),
        "Australia": (-43.7, -10.6, 113.3, 153.6),
        "Netherlands": (50.7, 53.6, 3.4, 7.2),
        "Germany": (47.2, 55.1, 5.8, 15.0),
        "Norway": (57.9, 71.2, 4.6, 31.1),
        "EIRE": (51.4, 55.4, -10.6, -5.3),
    }

    unique_id = df.CustomerID.dropna().unique()

    for i in unique_id:
        row_id = df[df["CustomerID"] == i].index
        CountryName = df.loc[df["CustomerID"] == i, "Country"].iloc[0]
        code = local.get(CountryName, "en_GB")
        fake = Faker(code)

        # Generate more realistic address
        street = fake.street_address()
        city = fake.city()
        postcode = fake.postcode()
        Address = f"{street}, {city}, {postcode}"

        CustomerName = fake.name()

        # Generate bounded lat/lon
        if CountryName in country_coords:
            lat_min, lat_max, lon_min, lon_max = country_coords[CountryName]
            Latitude = str(random.uniform(lat_min, lat_max))
            Longitude = str(random.uniform(lon_min, lon_max))
        else:
            Latitude = str(fake.latitude())
            Longitude = str(fake.longitude())

        for x in row_id:
            df.loc[x, "CustomerName"] = CustomerName
            df.loc[x, "Address"] = Address
            df.loc[x, "Latitude"] = Latitude
            df.loc[x, "Longitude"] = Longitude

    return df
```

```python
Faker.seed(5)

secure_db = anonymous(Ecommerce)
secure_db
```

```python
display_db = []
for i in Ecommerce.Country.dropna().unique():
    display_db.append(secure_db[secure_db["Country"] == i].to_numpy()[0])
pd.DataFrame(display_db, columns=Ecommerce.columns)
```
